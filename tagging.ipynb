Tag each word present in sample3.txt file with its POS using a pipeline containing BigramTagger --> UnigramTagger --> RegexpTagger --> DefaultTagger. Note that --> indicates backoff.

Use the data in the brown corpus as reference POS tag

```python
import nltk
from nltk import *
from nltk.corpus import brown
from string import punctuation
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import matplotlib.pyplot as plt
```

```python
input_file = "sample1.txt"
with open(input_file, 'r') as f_in:
    data = f_in.read()

tokens = word_tokenize(data)
postags = pos_tag(word_tokenize(brown))
```

```python
patterns = [
     (r'.*\'s$', 'NN$'),               # possessive nouns
     (r'.*es$', 'VBZ'),                # 3rd singular present
     (r'^-?[0-9]+(.[0-9]+)?$', 'CD'),  # cardinal numbers
     (r'[Aa][Nn][Dd]','CC'),           # conjugate and
     (r'.*ed$', 'VBD'),                # simple past
     (r',',','),                     # comma
     (r'.*ould$', 'MD'),               # modals
     (r'.*ing$', 'VBG'),               # gerunds
     (r'.*s$', 'NNS'),                 # plural nouns
    # You will need to add several such rules to make an efficient tagger
]
defaulttagger = nltk.DefaultTagger(tag='NN')
regexptagger = nltk.RegexpTagger(regexps=patterns, backoff=defaulttagger)
unigram = nltk.UnigramTagger(train = [postags], backoff=regexptagger)
bigram = nltk.NgramTagger(n=2, train=[postags], backoff=unigram)
tagging = bigram.tag(tokens)
tagging
```
